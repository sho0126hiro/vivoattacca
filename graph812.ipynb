```
#@title デフォルトのタイトル テキスト
#p133. fig8.12
import numpy as np
import matplotlib.pyplot as plt

for i in range(-4,4,2):
  if i in [-4,0] :plt.plot([i,i+2],[2,0],color='black')
  else :plt.plot([i,i+2],[0,2],color='black')
x=np.linspace(-4,4,100)
y=1
b=lambda n:(4*(1-np.cos(np.pi*n))/np.power(np.pi*n,2))

for n in range(1,10):
  y+=b(n)*np.cos(n*np.pi/2*x)
  plt.plot(x,y)
plt.show()
```

```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.font_manager

plt.figure( figsize=(15,6))
x= np.linspace(-4,4,100)

y=1
b=lambda n:(4*(1-np.cos(np.pi*n))/np.power(np.pi*n,2))

for n in range(1,6):
  y+=b(n)*np.cos(n*np.pi/2*x)
plt.plot(x,y)
plt.xlabel("t[s]")
plt.ylabel("f(t)")
plt.xticks([-4,-2,0,2,4])
plt.yticks(np.arange(-1,4,1))
plt.show()
```
